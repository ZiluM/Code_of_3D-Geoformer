```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dl_corr = np.load("./corr_dl.npy")
lim_corr = np.load("/glade/work/zilumeng/3D_trans/LIM/pred/lim_pred_nino34_corr.npy")
dl_corr,lim_corr
```

```python
months = np.arange(1,21)
plt.plot(months,dl_corr,'-o',label="DL",color="red")
plt.plot(months,lim_corr,'-o',label="LIM",color="blue")
plt.legend()
plt.xlabel("lead time (months)")
plt.ylabel("correlation")
plt.title("Nino3.4 prediction skill")
plt.xticks(months[::2])
plt.axhline(y=0.5,color="black",linestyle="--")
plt.savefig("./corr_nino34.png")
```
